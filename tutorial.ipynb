```python
import pleasingmatrixorder
import numpy as np
import pickle
import matplotlib.pylab as plt
%matplotlib inline
```

We start by loading a matrix

```python
with open('confusion.pkl','rb') as f:
    confusion=pickle.load(f)
```

Let's take a look at it

```python
plt.gcf().set_size_inches(10,10)
plt.imshow(confusion)
```

It is in an *unpleasing order*.  Let's use the hungarian algorithm to get it into a vaguely sensible order for most of the rows and columns, and then add back in the stragglers in places that seem to make sense.

```python
plt.gcf().set_size_inches(10,10)
posA,posB=pleasingmatrixorder.find_pleasing_order(confusion,mergify=False)
plt.imshow(confusion[np.ix_(posA,posB)])
```

That looks much better!  It is still kind of rough around the edges.  There are some funny guys way out away from the diagonal that I don't like.  Let us use the iterative partition merging to get a better order.

```python
posA2,posB2=pleasingmatrixorder.find_pleasing_order(confusion,mergify=True)
```

```python
plt.gcf().set_size_inches(10,10)
plt.imshow(confusion[np.ix_(posA2,posB2)])
```

That's better.
